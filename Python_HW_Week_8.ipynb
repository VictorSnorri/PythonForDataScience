3. Create an ECDF plot (*ecdfplot*) for **Grade**. What is the percentage of students scoring less than 85?



```python
proportion_below_85 = (df['Grade'] < 85).mean()

sns.ecdfplot(data=df, x='Grade') 
plt.axhline(y=proportion_below_85, color='red', linestyle='--', linewidth=1.5)

plt.xlabel('Grade', fontsize=12)
plt.ylabel('Proportion', fontsize=12)
plt.title('ECDF Plot for Grades', fontsize=14)
plt.legend()
plt.show()

plt.figure()  
sns.ecdfplot(data=df, x='Grade')  
plt.axhline(y=proportion_below_85, color='red', linestyle='--', linewidth=1.5, label=f'Proportion = {proportion_below_85:.3f}')
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Proportion', fontsize=12)
plt.title('Zoomed ECDF Plot for Grades', fontsize=14)
plt.legend()
plt.xlim(80, 90)  
plt.ylim(0.2, 0.4)  # 
plt.show()  
```

**Conclusion**: As seen in the second plot, the proportion of grades under 85 is 0.286.

5. Create a *swarmplot* to show the relationship between Gender and **Study Time**. Which gender has a higher average study time?



```python
sns.swarmplot(data=df, x='Gender', y='Study Time', palette='Set1')

plt.xlabel('Gender', fontsize=12)
plt.ylabel('Study Time (hours)', fontsize=12)
plt.title('Study Time by Gender', fontsize=14)

plt.show()
```
