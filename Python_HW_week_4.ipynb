```python
### WEEK 4 - Basic Libraries 

#Given a zip file with a subfolder with multiple annotations, where the name convention for each one of them is: 

#{DATE}_{TIME}_SN{SATELLITE_NUMBER}_QUICKVIEW_VISUAL_{VERSION}_{UNIQUE_REGION}.txt

#where:

#- DATE expressed as YYYYMMDD (year, month and day), e.g. 20241201, 20230321 ...
#- TIME expressed as HHMMSS (hour, minutes and seconds), e.g. 2134307
#- SATELLITE_NUMBER an integer that represents the satellite number.
#- VERSION provides the version of the pipeline, e.g. "0_1_2", "1_3_1" ...
#- UNIQUE_REGION provides a unique location in the form of a string, e.g SATL-2KM-10N_552_4164
```

```python
import pandas as pd
import glob
import os
```

```python
#1. How many files the annotations folder has.
len(os.listdir())
```

```python
#2. How many of them follow the name convention expressed above.
#create convention using wildcards
convention = '*_*_SN*_QUICKVIEW_VISUAL_*_*.txt'

#use glob.glob() to filter files
files_conv = glob.glob(convention)

#count files
len(files_conv)
```

```python
#3. How many of annotations you have per month and year. Which month has more annotation files.

#create year and month list
year = []
month = []

#loop through each filename in the filtered list to extract year and month
for file in files_conv:

    #extract the year 
    file_year = file[0:4]

    #append the extracted year to the year list
    year.append(file_year)

    #extract the month 
    file_month = file[4:6]

    #append the extracted month to the month list
    month.append(file_month)


#convert the list of years into a pandas Series and count the occurrences of each year
year = pd.Series(year)
month = pd.Series(month)

#count of files per year
year_count = year.value_counts()

#count of files per month
month_count = month.value_counts()

#print counts per year
print(year_count)
print('\n')

#print counts per month
print(month_count)
print('\n')

#the month with the most annotations
print(f'Month "{month_count.index[0]}" is the month with the most annotation-files, having "{month_count.iloc[0]}" files.')
```

```python
#5 Print all the annotations from the most recent to the oldest one. 
file_date = []

#ittare through converted files
for i in files_conv:
    #extract date
    i_date = i[0:14]

    #append date and file as a touple to file_date
    file_date.append((i_date, i))

#sort date time in descending order
file_date.sort(reverse= True)

#print just the file name
for d, f in file_date:
    print(f)

```

```python
#6. How many different satellites there are, how many annotations we have per satellite number, and which one was used in the most recent annotation file. 
#create a dictionary
satellite_dict = {}

#itterate through files
for i in files_conv:
    #get the satalite numbers
    satellite_nmbr = i[16:20]

    #if sattalite is not in dictionary already add and put value as 1
    if satellite_nmbr not in satellite_dict:
        satellite_dict[satellite_nmbr] = 1

    #else increase value by 1
    else:
        satellite_dict[satellite_nmbr] += 1


#print amount of satellites
print(f'There are {len(satellite_dict)} satellites')

#print every sattelite and their respective annotations
for i in satellite_dict:
    print(f'Sattelite {i} has {satellite_dict[i]} annotations')

#print the most recent annotation's satellite
print(f'{file_date[0][1][16:20]} was used in the most recent file')
        
```

```python
#7. How many unique regions there are.
import numpy as np
#create a regions list
regions = []

#itterate through files
for file in files_conv:
    #split using the split method
    parts = file.split('_')
    #grab unique regions by selecting the last part and deleting '.txt'
    d_region = parts[-1].replace('.txt', '')
    #append regions to list
    regions.append(d_region)

#create an array with unique regions
unique_regions = np.unique(regions)

#print the lenght
print(f' There are {len(unique_regions)} unique regions')

```
